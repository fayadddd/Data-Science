```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Population Data

```python
population_data =[ np.random.randint(1,100) for i in range(10000)]   # it will make(1 to 100)any random numbers 10,000 times.to make 10,000 population data
x_table = pd.DataFrame({"Population Data":population_data})          #  x = "Popuplation Data" , y = "Density"(automaticly)
```

```python
plt.figure(figsize=(4,3))
sns.kdeplot(x="Population Data",data=x_table)
plt.show()                                      # it show it's a (non normal distribution chart)
```

```python
np.mean(population_data)
```
