### ApplicantIncome Outlier Remove

```python
# data dhape
dataset.shape
```

```python
# Datatset [ApplicantIncome]

#q1 = 0.000000
q1 = dataset["ApplicantIncome"].quantile(0.25)
#q3 = 81000.000000
q3 = dataset["ApplicantIncome"].quantile(0.75)
```

```python
IQR = q3 -q1
```

```python
#minimum = 0.000000
minimum = q1 - (1.5*IQR)
#maximum = 41667.000000
maximum = q3 + (1.5*IQR)
```

```python
plt.figure(figsize=(15,5))
sns.boxplot(x = "CoapplicantIncome",data= dataset)
plt.show()
```

```python
# Remove Outlier
new_dataset = dataset[dataset["CoapplicantIncome"]<=maximum]
new_dataset.shape
```

```python
# final result
sns.boxplot(x = "CoapplicantIncome",data= new_dataset)
plt.show()
```
