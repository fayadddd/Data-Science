```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
```

# Standerlization Scaling

```python
# Standerlization Scaling = data's nature will be [same] but value will be [reduce]
```

```python
dataset = pd.read_csv(r"C:\Users\My\Downloads\loan_data_set.csv")
dataset.head()
```

```python
dataset.shape
```

```python
dataset.isnull().sum()
```

```python
dataset['ApplicantIncome'].describe()
```

```python
sns.distplot(dataset["ApplicantIncome"])
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
ss.fit(dataset[["ApplicantIncome"]])
```

```python
ss.transform(dataset[["ApplicantIncome"]])
```

```python
dataset["ApplicantIncome_ss"] = pd.DataFrame(ss.transform(dataset[["ApplicantIncome"]]),columns=["X"])
dataset.head()
```

```python
dataset.shape
```

```python
dataset.describe()
```

```python
plt.figure(figsize=(10,7))

plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(dataset["ApplicantIncome"])

plt.subplot(1,2,2)
plt.title("After")
sns.distplot(dataset["ApplicantIncome_ss"])
plt.show()               
```

```python
# same process to (CoapplicantIncome) & (Loan_Amount_Term)
```

# Normalization/min_max Scaling

```python
# Normalization/min_max Scaling = data's nature will be [same] but value will be [reduce in 0 to 1]
```

```python
dataset = pd.read_csv(r"C:\Users\My\Downloads\loan_data_set.csv")
dataset.head()
```

```python
dataset.shape
```

```python
dataset.isnull().sum()
```

```python
dataset["CoapplicantIncome"].describe()
```

```python
sns.distplot(dataset["CoapplicantIncome"])
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mms = MinMaxScaler()
mms.fit(dataset[["CoapplicantIncome"]])
```

```python
mms.transform(dataset[["CoapplicantIncome"]])
```

```python
dataset["CoapplicantIncome_mms"] = mms.transform(dataset[["CoapplicantIncome"]])
dataset.head()
```

```python
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(dataset["CoapplicantIncome"])

plt.subplot(1,2,2)
plt.title("After")
sns.distplot(dataset["CoapplicantIncome_mms"])
plt.show() 
```

```python
# same process to (ApplicantIncome) & (Loan_Amount_Term)
```
