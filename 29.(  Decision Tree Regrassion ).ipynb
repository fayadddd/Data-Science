```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions
```

```python
dataset = pd.read_csv(r"C:\Users\My\Downloads\EmployeeText.csv")
dataset = pd.read_csv(r"C:\Users\My\Downloads\EmployeeText.csv", thousands=',')
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
sns.pairplot(data=dataset)
plt.show()
```

```python
x = dataset.iloc[:,:-1]
y = dataset["  Experience"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
# Decision Tree Regressor Model
dt = DecisionTreeRegressor()
dt.fit(x_train,y_train)
```

```python
dt.score(x_test,y_test)*100
```

```python
dt.score(x_train,y_train)*100
```

## Post Pruning

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
for i in range(1,20):
    dt = DecisionTreeClassifier(max_depth=i)
    dt.fit(x_train,y_train)
    print(dt.score(x_train,y_train)*100,dt.score(x_test,y_test)*100,i)
```

```python
dt = DecisionTreeClassifier(max_depth=1)
dt.fit(x_train,y_train)
```

```python
dt.score(x_train,y_train)*100
```

```python
dt.score(x_test,y_test)*100
```

```python
from sklearn.tree import plot_tree
```

```python
plot_tree(dt)
plt.show()
```
