```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.datasets import make_moons
```

```python
make_moons(n_samples=250,noise=0.05)
```

```python
x,y = make_moons(n_samples=250,noise=0.05)
```

```python
df = {"data1" : x[:,0] , "data2": x[:,1]}
```

```python
dataset = pd.DataFrame(df)
dataset.head()
```

```python
sns.scatterplot(x="data1", y="data2", data=dataset)
plt.show()
```

# DBSCAN Model

```python
from sklearn.cluster import DBSCAN
```

```python
db = DBSCAN(eps=0.3,min_samples=5)        # eps changeing for best cluster tuning (0.5) in idle
db.fit_predict(dataset) 
```

```python
dataset["predict"]=db.fit_predict(dataset)
dataset.head()
```

## Number of Clusters

```python
sns.scatterplot(x="data1", y="data2", data=dataset, hue="predict")
plt.show()
```

## Original Number of Clusters

```python
sns.scatterplot(x="data1", y="data2", data=dataset, hue=y)
plt.show()
```
