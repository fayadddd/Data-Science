```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.datasets import make_moons
```

```python
x,y = make_moons(n_samples=1000,noise=0.2)
```

```python
df = {"x1":x[:,0],"x2":x[:,1],"y":y}
```

```python
dataset = pd.DataFrame(df)
dataset.head()
```

```python
sns.scatterplot(x="x1",y="x2",data=dataset,hue="y")
plt.show()
```

```python
x_a = dataset.iloc[:,:-1]
y_a = dataset["y"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_a, y_a, test_size=0.2, random_state=42)
```

# Meta-Estimator

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
```

```python
bg = BaggingClassifier(estimator=SVC(),n_estimators=50)
bg.fit(x_train,y_train)
```

```python
bg.score(x_train, y_train)*100, bg.score(x_test,y_test)*100
```

```python
bg_1 = BaggingClassifier(estimator=RandomForestClassifier(),n_estimators=50)
bg_1.fit(x_train,y_train)
```

```python
bg_1.score(x_train, y_train)*100, bg_1.score(x_test,y_test)*100
```

```python
bg_2 = BaggingClassifier(estimator=GaussianNB(),n_estimators=50)
bg_2.fit(x_train,y_train)
```

```python
bg_2.score(x_train, y_train)*100, bg_2.score(x_test,y_test)*100
```

# Random Forest

```python
rf = RandomForestClassifier(n_estimators=80)
rf.fit(x_train,y_train)
```

```python
rf.score(x_train, y_train)*100, rf.score(x_test,y_test)*100
```
