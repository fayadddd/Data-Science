```python
!pip install imblearn
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
dataset = pd.read_csv(r"C:\Users\My\Downloads\Social_Network_Ads.csv")
dataset.head()
```

```python
dataset["Purchased"].value_counts()
```

```python
x = dataset.iloc[:,:-1]
y = dataset["Purchased"]
```

# Random Under Sample

```python
from imblearn.under_sampling import RandomUnderSampler
```

```python
rus = RandomUnderSampler()
rus_x , rus_y = rus.fit_resample(x,y)
```

```python
rus_y.value_counts()
```

```python
x_train, x_test, y_train, y_test = train_test_split(rus_x,rus_y,test_size=0.2,random_state=42)
```

```python
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
# test-1
lr.predict([[19,19000]])
```

```python
# test-2
lr.predict([[47,49000]])
```

```python
# test-3
lr.predict([[34,112000]])
```

# Random Over Sample

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
ros = RandomOverSampler()
ros_x , ros_y = ros.fit_resample(x,y)
```

```python
ros_y.value_counts()
```

```python
x_train, x_test, y_train, y_test = train_test_split(rus_x,rus_y,test_size=0.2,random_state=42)
```

```python
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
# test-1
lr.predict([[19,19000]])
```

```python
# test-2
lr.predict([[47,49000]])
```
