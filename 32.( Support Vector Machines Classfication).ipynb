```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions
```

```python
dataset = pd.read_csv(r"C:\Users\My\Downloads\placement (1).csv")
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
plt.figure(figsize=(5,4))
sns.scatterplot(x ="cgpa", y="resume_score",data=dataset,hue="placed")
plt.show()
```

```python
x = dataset.iloc[:,:-1]
y = dataset["placed"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25,random_state=15)
```

# Support Vector Machine

```python
from sklearn.svm import SVC
```

## Linear SVM Model

```python
sv = SVC (kernel="linear")
sv.fit(x_train,y_train)
```

```python
sv.score(x_test,y_test)*100
```

```python
sv.score(x_train,y_train)*100
```

```python
plt.figure(figsize=(5,4))
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf = sv)
plt.show()
```

## poly SVM Model

```python
sv = SVC (kernel="poly")
sv.fit(x_train,y_train)
```

```python
sv.score(x_test,y_test)*100
```

```python
plt.figure(figsize=(5,4))
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf = sv)
plt.show()
```

## rbf SVM Model

```python
sv = SVC (kernel="rbf")
sv.fit(x_train,y_train)
```

```python
sv.score(x_test,y_test)*100
```

```python
sv.score(x_train,y_train)*100
```

```python
plt.figure(figsize=(5,4))
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf = sv)
plt.show()
```

## sigmoid SVM Model

```python
sv = SVC (kernel="sigmoid")
sv.fit(x_train,y_train)
```

```python
sv.score(x_test,y_test)*100
```

```python
plt.figure(figsize=(5,4))
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf = sv)
plt.show()
```
