```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv(r"C:\Users\My\Downloads\iris.csv")
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
sns.pairplot(data=dataset)
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []

for i in range(2,21): 
    km = KMeans(n_clusters=i, init='k-means++')
    km.fit(dataset)
    wcss.append(km.inertia_)
```

```python
plt.figure(figsize=(10,5))
plt.plot([i for i in range(2,21)],wcss,marker="o")
plt.xticks([i for i in range(2,21)])
plt.xlabel("no of Cluster")
plt.ylabel("WCSS")
plt.grid('x')
plt.show()
```

```python
# chose 3 no_of_Cluster cz the graph fast break on 3 no_of_Cluster
```

```python
kmn = KMeans(n_clusters=3)
kmn.fit_predict(dataset)
```

```python
dataset["Predicted_Data"] = kmn.fit_predict(dataset)
```

```python
dataset.head()
```

```python
sns.pairplot(data=dataset, hue="Predicted_Data")
plt.show()
```
