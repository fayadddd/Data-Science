# Logistic Regression

## Single Binary classification

```python
dataset = pd.read_csv(r"C:\Users\My\Downloads\Social_Network_Ads.csv")
dataset.head()
```

```python
dataset.drop(columns=["User ID","Gender","EstimatedSalary"],inplace=True)
dataset.head()
```

```python
plt.figure(figsize=(5,4))
sns.scatterplot(x="Age", y="Purchased",data=dataset)
plt.show()
```

```python
x = dataset[["Age"]]
y = dataset["Purchased"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
# prediction-1
lr.predict([[40]])
```

```python
# prediction-2
lr.predict([[19]])
```

```python
# prediction-3
lr.predict([[43]])
```

```python
# prediction-4
lr.predict([[50]])
```

```python
plt.figure(figsize=(5,4))
sns.scatterplot(x="Age", y="Purchased",data=dataset)
sns.lineplot(x="Age", y=lr.predict(x),data=dataset,c="red")
plt.show()
```

## Polynomial Binary classification

```python
dataset = pd.read_csv(r"C:\Users\My\Downloads\Polynomial Classification.csv")
dataset.head()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x="data-1",y="data-2", data=dataset, hue="output")
plt.show()
```

```python
x = dataset.iloc[:,:-1]
y = dataset["output"]
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
# Polynomial Model
pf = PolynomialFeatures(degree=3)
pf.fit(x)
x = pd.DataFrame(pf.transform(x))
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

## Mluticlass Classification 

```python
dataset = pd.read_csv(r"C:\Users\My\Downloads\iris.csv")
dataset.head()
```

```python
dataset["variety"].unique()
```

```python
sns.pairplot(data=dataset,hue="variety")
plt.show()
```

```python
# Fiture selection
# at the bottom, 
# row_1 and column_4 show the petal.width graph is Good ( 3 defferent line are not mixed togather ) 
# row_2 and column_3 show the petal.width graph is Good ( 3 defferent line are not mixed togather ) 
# row_3 and column_2 show the petal.width graph is veryBad ( 3 defferent line are mixed togather )
# row_4 and column_1 show the petal.width graph is Bad ( 3 defferent line are mixed togather )
```

```python
x = dataset.iloc[:,:-1]
y = dataset["variety"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
# OVR Method
lr = LogisticRegression(multi_class='ovr')
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
# Multinomial Method
lr = LogisticRegression(multi_class='multinomial')
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
# Normal Method
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```
