```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions
```

```python
dataset = pd.read_csv(r"C:\Users\My\Downloads\placement (1).csv")
dataset.head()
```

```python
plt.figure(figsize=(4,3))
sns.kdeplot(data=dataset["cgpa"])
plt.title("semi Normal Dristibution")
plt.show()
# semi normal Dristibution
```

```python
plt.figure(figsize=(4,3))
sns.kdeplot(data=dataset["resume_score"])
plt.title("Normal Dristibution")
plt.show()
# normal Dristibution
```

```python
plt.figure(figsize=(4,3))
sns.kdeplot(data=dataset["placed"])
plt.title(" None Normal Dristibution")
plt.show()
# None normal Dristibution
```

```python
plt.figure(figsize=(4,3))
sns.scatterplot(x="cgpa",y="resume_score",data=dataset,hue="placed")
plt.show()
```

```python
x = dataset.iloc[:,:-1]
y = dataset["placed"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

# Gaussian

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb = GaussianNB()
gnb.fit(x_train,y_train)
```

```python
gnb.score(x_test,y_test)*100 , gnb.score(x_train,y_train)*100
```

```python
plt.figure(figsize=(4,3))
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=gnb)
plt.show()
```

# Multinomial

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
mln = MultinomialNB()
mln.fit(x_train,y_train)
```

```python
mln.score(x_test,y_test)*100 , mln.score(x_train,y_train)*100
```

```python
plt.figure(figsize=(4,3))
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=mln)
plt.show()
```

# Bernoulli

```python
from sklearn.naive_bayes import BernoulliNB
```

```python
bn = BernoulliNB()
bn.fit(x_train,y_train)
```

```python
bn.score(x_test,y_test)*100 , bn.score(x_train,y_train)*100
```

```python
plt.figure(figsize=(4,3))
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=bn)
plt.show()
```
