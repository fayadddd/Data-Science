# Ordinal Encoding

```python
# Ordinal Encoding is orde rwise/choise wise Encode
```

```python
#--1
df = pd.DataFrame({"Size":["s","m","l","xl","l","s","s","m","xl"]})
df.head()
```

```python
data = [["s","m","l","xl"]]              # dual dimantional that's why [[]]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
ordinal_Encoding = OrdinalEncoder(categories=data)
ordinal_Encoding.fit(df[["Size"]])
```

```python
ordinal_Encoding.transform(df[["Size"]])
```

```python
df["Size_Encode"] = ordinal_Encoding.transform(df[["Size"]])
df
```

## Map Funtion

```python
#--2
df = pd.DataFrame({"Size":["s","m","l","xl","l","s","s","m","xl"]})
df.head()
```

```python
data = {"s":0,"m":1,"l":2,"xl":3}
```

```python
df["Size"].map(data)
```

```python
df["Size_map_Encode"] = df["Size"].map(data)
df
```
