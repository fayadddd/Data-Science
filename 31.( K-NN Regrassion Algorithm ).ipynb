```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions
```

```python
dataset = pd.read_csv(r"C:\Users\My\Downloads\EmployeeText.csv")
dataset = pd.read_csv(r"C:\Users\My\Downloads\EmployeeText.csv", thousands=',')
dataset.head()
```

```python
x = dataset.drop(columns="  Experience")
y = dataset["  Experience"]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
sc.fit(x)
x = pd.DataFrame(sc.transform(x),columns=x.columns)
x.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
for i in range(1,30):
    knn = KNeighborsRegressor (n_neighbors=i)
    knn.fit(x_train,y_train)
    print(knn.score(x_train,y_train)*100,knn.score(x_test,y_test)*100,i)
```

```python
knn = KNeighborsRegressor(n_neighbors = 29)
knn.fit(x_train,y_train)
```
