```python
import pandas as pd
from mlxtend.feature_selection import SequentialFeatureSelector
```

```python
dataset = pd.read_csv(r"C:\Users\My\Downloads\diabetes.csv")
dataset.head()
```

```python
# seperate X new column and Y new column
x = dataset.iloc[:,:-1]        # iloc[row,column]    # iloc[all_row,n-1]
y = dataset["Outcome"]  
```

```python
x.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
Logis_Reg = LogisticRegression()
```

# Froword Selection


## Forword Feature_selection

```python
fs = SequentialFeatureSelector(Logis_Reg, k_features=8, forward=True)
fs.fit(x,y)
```

```python
fs.feature_names
```

```python
fs.k_feature_names_
```

```python
eight = fs.k_score_
eight
```
