## Linkgae

### Single Linkage

```python
import scipy.cluster.hierarchy as sc
```

```python
sc.dendrogram(sc.linkage(dataset, method='single', metric='euclidean'))
plt.show()
```

### Complete Linkage

```python
sc.dendrogram(sc.linkage(dataset, method='complete', metric='euclidean'))
plt.show()
```

### Average Linkage

```python
sc.dendrogram(sc.linkage(dataset, method='average', metric='euclidean'))
plt.show()
```

### Centroid Linkage

```python
sc.dendrogram(sc.linkage(dataset, method='centroid', metric='euclidean'))
plt.show()
```

## Cluster Model

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
#   ?????????????
ac = AgglomerativeClustering(n_clusters=2,linkage="single")
dataset["Predicted_data"] = ac.fit_predict(dataset)
dataset.head()
```

```python
sns.pairplot(data=dataset,hue="Predicted_data")
plt.show()
```
